```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
sns.set_style('whitegrid')
```

```python
iris = load_iris()
```

```python
X = iris.data
```

```python
y = iris.target
```

```python
X
```

```python
y
```

```python
# Variance
# going to remove low variance features
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
sel = VarianceThreshold(threshold=0.5)
```

```python
X_vt = sel.fit_transform(X)
```

```python
X_vt
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(X_vt[:,0],X_vt[:,1],X_vt[:,2],c=y,cmap='rainbow')
plt.show()
```

```python
sel = VarianceThreshold(threshold=0.6)
```

```python
X_vt = sel.fit_transform(X)
```

```python
X_vt
```

```python
plt.scatter(X_vt[:,0],X_vt[:,1],c=y,cmap='rainbow')
plt.show()
```

```python
# Select from model
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.feature_selection import SelectFromModel
```

```python
clf = SVC(kernel='linear')
```

```python
clf.fit(X,y)
```

```python
model = SelectFromModel(clf,prefit=True)
```

```python
X_sfm = model.transform(X)
```

```python
X_sfm
```

```python
plt.scatter(X_sfm[:,0],X_sfm[:,1],c=y,cmap='rainbow')
plt.show()
```
